```python
df["Order Date"]=pd.to_datetime(df["Ship Date"], format='%d/%m/%Y')
```

```python
df["Year"]=df["Order Date"].dt.year
df
```

```python
df["Year"].value_counts()
```

```python
#hangi ayda en fazla satış yapılmış 
df["Month"].value_counts()
```

```python
# hangi ayda satış fazla
months_df= df.groupby("Month").sum(numeric_only=True)
months_df
```

```python
months_df=months_df[["Sales"]]
months_df
```

```python
months_df.reset_index(inplace=True)
```

```python
months_df
```

```python
#satış tutarlarını görselleştirme
plt.bar(months_df["Month"],months_df["Sales"])
plt.xlabel("Aylar")
plt.ylabel("Satış Miktarı")
plt.xticks(fontsize=14,rotation=90)
plt.yticks(fontsize=14)
plt.show()
```

```python
# yıllar için
year_df=df.groupby("Year").sum(numeric_only=True)
year_df=year_df[["Sales"]]
year_df
```

```python
year_df.reset_index(inplace=True)
```

```python
plt.bar(year_df["Year"],year_df["Sales"])
plt.xlabel("Yıllar")
plt.ylabel("Satış Miktarı")
plt.xticks(fontsize=14,rotation=90)
plt.yticks(fontsize=14)
plt.show()
```

```python
# en çok hangi müşteri para harcamış

customer=df.groupby("Customer Name").sum(numeric_only=True).sort_values("Sales",ascending=False)
customer.drop(["Postal Code","Year"],axis=1,inplace=True)
customer
```

```python
customer.reset_index(inplace=True)
```

```python
customer
```

```python
customer=customer.head(20)
```

```python
plt.bar(customer["Customer Name"],customer["Sales"])
plt.xlabel("Kişi")
plt.ylabel("Satış ")
plt.xticks(fontsize=14,rotation=90)
plt.yticks(fontsize=14)
plt.show()
```

```python
# kategoriler üzerinden satış miktarı

category= df.groupby("Category").sum(numeric_only=True)
category.drop(["Postal Code","Year"],axis=1,inplace=True)
category
```

```python
category.reset_index(inplace=True)
category
```

```python
plt.bar(category["Category"],category["Sales"])
plt.xlabel("Kategori")
plt.ylabel("Satış ")
plt.xticks(fontsize=14,rotation=90)
plt.yticks(fontsize=14)
plt.show()
```

```python
plt.pie(category["Sales"],labels=category["Category"],autopct="%0.2f%%")
plt.show()
```
