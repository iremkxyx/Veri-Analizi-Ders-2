```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\Lenovo\Veri-Analizi-2\datasets\Student_Grades.csv")
```

```python
df.head(3)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
y=df["Scores"]
x=df.drop(columns=["Scores","Grade"])

y=y.values.astype("float32")
x=x.values.astype("float32")
```

```python
model=Sequential()

model.add(Dense(16))
model.add(Dense(32,activation="relu"))
model.add(Dense(32,activation="relu"))
model.add(Dense(1)) #çıkış

model.compile(optimizer="adam",loss="mse")

model.fit(x,y,epochs=128,batch_size=16)
```

```python
model.predict(np.array([[3,1,1,2,3.3]]))
```
